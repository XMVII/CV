# **Q2**

## **1) RGB 이미지 출력**

```python
import cv2

from google.colab.patches import cv2_imshow

path = "/content/drive/MyDrive/My Drive/lena.jpg"
image = cv2.imread(path, cv2.IMREAD_COLOR)

cv2_imshow(image)
```

## **2, 3, 4) R, G, B-channel 출력**

```python
import matplotlib.pyplot as plt

RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
R_img, G_img, B_img = cv2.split(RGB_image)
plt.subplot(1, 4, 1)
plt.title("Original")
plt.imshow(RGB_image)

RGB_image[:,:,0] = R_img
RGB_image[:,:,1] = 0
RGB_image[:,:,2] = 0
plt.subplot(1, 4, 2)
plt.title("Red")
plt.imshow(RGB_image)

RGB_image[:,:,0] = 0
RGB_image[:,:,1] = G_img
RGB_image[:,:,2] = 0
plt.subplot(1, 4, 3)
plt.title("Green")
plt.imshow(RGB_image)

RGB_image[:,:,0] = 0
RGB_image[:,:,1] = 0
RGB_image[:,:,2] = B_img
plt.subplot(1, 4, 4)
plt.title("Blue")
plt.imshow(RGB_image)
```

## **5) Gray scale 이미지로 변환**

```python
Gray_image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

cv2_imshow(Gray_image)
```

## **6) Binary 이미지로 변환**

```python
imageGray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, dst = cv2.threshold(imageGray, 100, 255, cv2.THRESH_BINARY)

cv2_imshow(dst)
```

## **7, 8, 9) 5)번 작업을 수행한 후 출력 이미지에서 최대, 최소, 평균 화소값 출력**

```python
import cv2
import numpy as np

# 이미지 파일 경로 설정
image_path = "/content/drive/MyDrive/CV_Practice_Images/lena.jpg"

# 이미지 파일 읽기 (gray scale)
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 최대, 최소 화소값 계산
max_value = np.max(img)
min_value = np.min(img)

# 평균 화소값 계산
mean_value = np.mean(img)

# 계산된 값 출력
print("Max pixel value:", max_value)
print("Min pixel value:", min_value)
print("Mean pixel value:", mean_value)
```
