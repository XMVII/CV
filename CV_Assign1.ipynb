## **1) RGB 이미지 출력**

```python
import cv2

from google.colab.patches import cv2_imshow

path = "/content/drive/MyDrive/My Drive/lena.jpg"
image = cv2.imread(path, cv2.IMREAD_COLOR)

cv2_imshow(image)
```

## **2, 3, 4) R, G, B-channel 출력**

```python
import matplotlib.pyplot as plt

RGB_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
R_img, G_img, B_img = cv2.split(RGB_image)
plt.subplot(1, 4, 1)
plt.title("Original")
plt.imshow(RGB_image)

RGB_image[:,:,0] = R_img
RGB_image[:,:,1] = 0
RGB_image[:,:,2] = 0
plt.subplot(1, 4, 2)
plt.title("Red")
plt.imshow(RGB_image)

RGB_image[:,:,0] = 0
RGB_image[:,:,1] = G_img
RGB_image[:,:,2] = 0
plt.subplot(1, 4, 3)
plt.title("Green")
plt.imshow(RGB_image)

RGB_image[:,:,0] = 0
RGB_image[:,:,1] = 0
RGB_image[:,:,2] = B_img
plt.subplot(1, 4, 4)
plt.title("Blue")
plt.imshow(RGB_image)
```

## **7, 8, 9) 5)번 작업을 수행한 후 출력 이미지에서 최대, 최소, 평균 화소값 출력**

```python
import cv2
import numpy as np

# 이미지 파일 경로 설정
image_path = "/content/drive/MyDrive/CV_Practice_Images/lena.jpg"

# 이미지 파일 읽기 (gray scale)
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 최대, 최소 화소값 계산
max_value = np.max(img)
min_value = np.min(img)

# 평균 화소값 계산
mean_value = np.mean(img)

# 계산된 값 출력
print("Max pixel value:", max_value)
print("Min pixel value:", min_value)
print("Mean pixel value:", mean_value)
```

## **2-b) Wrapping add**

```python
import cv2
import numpy as np

def add_wrap_gray(img1, img2):
    # 입력 이미지와 같은 크기의 빈 이미지 생성
    output = np.zeros_like(img1)
    
    # 두 이미지 더하기
    sum_img = img1 + img2
    
    # 0 ~ 255 범위를 넘어가면 wrapping 방식으로 처리
    output[sum_img > 255] = sum_img[sum_img > 255] - 256
    output[sum_img < 0] = sum_img[sum_img < 0] + 256
    
    # 결과 이미지 반환
    return output

# 이미지 파일 경로 설정
image_path1 = "/content/drive/MyDrive/CV_Practice_Images/lena.jpg"
image_path2 = "/content/drive/MyDrive/CV_Practice_Images/mandrill.jpg"

# 이미지 파일 읽기 (gray scale로 변환)
image1 = cv2.cvtColor(cv2.imread(image_path1), cv2.COLOR_BGR2GRAY)
image2 = cv2.cvtColor(cv2.imread(image_path2), cv2.COLOR_BGR2GRAY)

# 이미지 파일 더하기 (wrapping 방식)
result = add_wrap_gray(image1, image2)

from google.colab.patches import cv2_imshow
cv2_imshow(result)
```
